###4. Data Visualization

```python
def avg_un():
    avg = {}
    
    for c in range(1991, 2022):
        data_value = df[f"{c}"].mean()
        avg[f"{c}"] = data_value
        
    return avg
```

```python
def country_rates():
    result = {}
    for i in range(0,235):
        country_rate = df.values[i][2:].mean()
        country_name = df.values[i][:1]
        result[f"{country_name}"] = country_rate
    return result
```

```python
country_rates()
```
