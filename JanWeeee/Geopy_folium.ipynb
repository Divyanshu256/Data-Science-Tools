Task 2: Consider Names of two places from the user and find out the distance between them using
Geopy library and Mark the location using folium.

```python
from geopy.distance import geodesic as GD   
from geopy.geocoders import Nominatim
```

```python
loc1 = input("Enter Location1 : ")
loc2 = input("Enter Location2 : ")
world=folium.Map()
geolocator = Nominatim(user_agent="my_raequest")
location1 = geolocator.geocode(loc1)
location2 = geolocator.geocode(loc2)
la1 , lo1 = location1.latitude, location1.longitude;
la2 , lo2 = location2.latitude, location2.longitude;
points = []
place_lat=[la1,la2]
place_lng=[lo1,lo2]
for i in range(len(place_lat)):
    points.append([place_lat[i], place_lng[i]])

city1 = (la1,lo1)  
city2 = (la2,lo2)  
for index,lat in enumerate(place_lat):
    folium.Marker([lat, 
                   place_lng[index]],).add_to(world)
folium.PolyLine(points, color='red').add_to(world)
print("The distance between "+loc1+" and "+loc2+" is: ", GD(city1,city2).km) 
world
```
