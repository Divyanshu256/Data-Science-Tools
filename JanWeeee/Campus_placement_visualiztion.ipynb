Since we have very less number of outliers and have majority of the outliers in hsc percentage so we can clear them.

```python
Q1 = df['hsc_p'].quantile(0.25)
Q3 = df['hsc_p'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (df['hsc_p'] >= Q1 - 1.5 * IQR) & (df['hsc_p'] <= Q3 + 1.5 *IQR)
placement_filtered=df.loc[filter]
```

```python
plt.figure(figsize = (15, 5))
plt.style.use('seaborn-white')
ax=plt.subplot(121)
plt.boxplot(df['hsc_p'])
ax.set_title('Before removing outliers(hsc_p)')
ax=plt.subplot(122)
plt.boxplot(placement_filtered['hsc_p'])
ax.set_title('After removing outliers(hsc_p)')
```

```python
#Checking placement status
plt.figure(figsize = (15,5))
explode = [0,0.15]
plt.subplot(121)
df['status'].value_counts().plot(kind ='pie',autopct = '%.2f',explode =explode)
plt.subplot(122)
sns.countplot(data = df, x = 'status', hue = df['gender'])
```

More male students are placed than female students 

```python
plt.figure(figsize = (15,5))
map= pd.crosstab([df['gender'] , df['status']],[df['specialisation'],df['degree_t'],df['hsc_s']], normalize = 'index')
sns.heatmap(map)
```
