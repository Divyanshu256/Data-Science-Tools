2) Replace all abnormal symbols such as ‘?’ ‘***’ etc. by Null values

```python
D1=dataset.replace('??',np.nan)
D2=D1.replace('????',np.nan)
display(D2)
```

3.Count the total number of null values present in each attribute (column)

```python
price=D2['Price'].isnull().sum()
print(price)
age=D2['Age'].isnull().sum()
print(age)
km=D2['KM'].isnull().sum()
print(km)
fuel=D2['FuelType'].isnull().sum()
print(fuel)
color=D2['MetColor'].isnull().sum()
print(color)
auto=D2['Automatic'].isnull().sum()
print(auto)
cc=D2['CC'].isnull().sum()
print(cc)
doors=D2['Doors'].isnull().sum()
print(doors)
weight=D2['Weight'].isnull().sum()
print(weight)
```

4) Replace the NaN in Age column by median value.

```python
x = dataset['Age'].median()
x
```

```python
dataset['Age'].fillna((dataset['Age'].median()),inplace=True)
dataset 
```

5) Replace the string value in column Door with number.

```python
dataset['Doors'].value_counts()
```

```python
dataset.replace({'Doors':{'three': 3,'four': 4, 'five' : 5}},inplace=True)
dataset['Doors'].value_counts()
```

6) Returns a Series containing counts of unique values for column ‘FuelType’ (Petrol,
Diesel, CNG )

```python
fueltype=dataset['FuelType'].value_counts(ascending=True)
display(fueltype)
```

7) Get the mode value of FuelType and fill NA/NaN values using the specified value of mode

```python
a = dataset['FuelType'].mode()
a
```

```python
dataset['FuelType'].fillna(dataset['FuelType'].mode()[0], inplace=True)
display(dataset)
```

8) Get the mode value of Metcolor and fill NA/NaN values using the specified value of mode.

```python
y = dataset['MetColor'].mode()
y
```

```python
dataset['MetColor'].fillna(dataset['MetColor'].mode()[0], inplace=True)
display(dataset)
```

9) Replace the value of FuelType as for Petrol replaces with 0, diesel with 1 and CNG with 2.

```python
dataset["FuelType"].replace({"Petrol": "0", "Diesel": "1","CNG":"2"}, inplace=True)
display(dataset)
```
